```python
import torch
import cv2              
import os      
import numpy as np
from torch.utils.data import Dataset 
import matplotlib.pyplot as plt  
from torch.utils.tensorboard import SummaryWriter

writer = SummaryWriter("logs")

# 【修正1】路径前加 r，或者把 \ 改成 /
# 还要确保这个路径下真的有这张图！建议用绝对路径如果不确定相对路径的话
image_path = r"data\train\ants_image\0013035.jpg"

# 读取图片
img_cv = cv2.imread(image_path)

# 【调试技巧】先检查一下是不是读到了，防止报错
if img_cv is None:
    print("错误：没找到图片！请检查 image_path 路径是否完全正确。")
else:
    # 颜色转换 BGR -> RGB
    img_cv = cv2.cvtColor(img_cv, cv2.COLOR_BGR2RGB)
    img_array = np.array(img_cv)

    # 【修正2】是 add_image 不是 add.image
    # 【修正3】加上 dataformats='HWC'，因为 OpenCV 读出来是 Height, Width, Channel
    writer.add_image("test", img_array, 1, dataformats='HWC')
    print("图片添加成功！")

# y = x 标量记录
for i in range(100):
    writer.add_scalar("y=x", i, i)

writer.close()
```
